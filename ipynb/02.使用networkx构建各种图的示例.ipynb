# 02.使用networkx构建各种图的示例

```python
import networkx as nx

print(nx.__version__)
```

## 无向图

```python
# 创建一个无向图
G = nx.Graph()

# 添加节点
G.add_nodes_from(['A', 'B', 'C', 'D', 'E'])

#添加边
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'C'), ('B', 'D'), ('C', 'D'), ('D', 'E')])

print(nx.is_directed(G))
print(nx.is_weighted(G))
```

```python
# 绘制图
nx.draw(G, with_labels=True, pos=nx.spring_layout(G))
```

```python
print(G.degree())  # 打印每个节点的度
print(G.nodes())  # 打印所有节点
print(G.edges())  # 打印所有边
print(G.number_of_nodes())  # 打印节点数
print(G.number_of_edges())  # 打印边数
```

```python
# 邻接矩阵
A = nx.adjacency_matrix(G)
print(A)

print(A.todense())
```

```python
 # 度矩阵
import numpy as np
 
D = np.diag(list(dict(nx.degree(G)).values()))
print(D)

# 拉普拉斯矩阵
L = D - A
print(L)

# 关联矩阵
B = nx.incidence_matrix(G)
print(B)

print(B.todense())
```

```python
# 节点的邻居
print(list(G.neighbors('A')))
print(list(G.neighbors('B')))
print(list(G.neighbors('C')))
print(list(G.neighbors('D')))
print(list(G.neighbors('E')))
```

## 有向图

```python
# 创建一个有向图
G = nx.DiGraph()

# 添加节点
G.add_nodes_from(['A', 'B', 'C', 'D', 'E'])

#添加边
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'C'), ('B', 'D'), ('C', 'D'), ('D', 'B'), ('D', 'E')])

print(nx.is_directed(G))
print(nx.is_weighted(G))
# 绘制图
nx.draw(G, with_labels=True, pos=nx.spring_layout(G))

```

```python
print(G.degree())  # 打印每个节点的度
print(G.nodes())  # 打印所有节点
print(G.edges())  # 打印所有边
print(G.number_of_nodes())  # 打印节点数
print(G.number_of_edges())  # 打印边数
print(G.in_degree())  # 打印每个节点的入度
print(G.out_degree())  # 打印每个节点的出度
```

```python
# 邻接矩阵
A = nx.adjacency_matrix(G)
print(A)

print(A.todense())
 # 度矩阵
D = np.diag(list(dict(nx.degree(G)).values()))
print(D)

# 拉普拉斯矩阵
L = D - A
print(L)

# 关联矩阵
B = nx.incidence_matrix(G)
print(B)

print(B.todense())
```

```python
print(list(G.neighbors('A')))
print(list(G.neighbors('B')))
print(list(G.neighbors('C')))
print(list(G.neighbors('D')))
print(list(G.neighbors('E')))
```

## 赋权图

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 创建一个赋权图
G = nx.Graph()

# 添加节点
G.add_nodes_from(['A', 'B', 'C', 'D', 'E'])

# 添加边
G.add_edges_from([('A', 'B', {'weight': 3}), ('A', 'C', {'weight': 2}), ('B', 'C', {'weight': 4}), ('B', 'D', {'weight': 3}), ('C', 'D', {'weight': 1}), ('D', 'E', {'weight': 5})])

# 绘制图形
pos = nx.spring_layout(G)  # 选择一个布局算法
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=10, font_weight='bold', font_color='black')

# 添加权重标签
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, font_size=8, font_color='red')

plt.show()
```

```python
print(nx.is_directed(G))
print(nx.is_weighted(G))
```

```python
print(G.degree())  # 打印每个节点的度
print(G.nodes())  # 打印所有节点
print(G.edges())  # 打印所有边
print(G.number_of_nodes())  # 打印节点数
print(G.number_of_edges())  # 打印边数
print(G.get_edge_data('A', 'B'))
print(G.get_edge_data('A', 'C'))
print(G.get_edge_data('B', 'C'))
print(G.get_edge_data('B', 'D'))
print(G.get_edge_data('C', 'D'))
print(G.get_edge_data('D', 'E'))
```

```python
# 邻接矩阵
A = nx.adjacency_matrix(G)
print(A)

print(A.todense())

# 度矩阵
D = np.diag(list(dict(nx.degree(G)).values()))
print(D)

# 拉普拉斯矩阵
L = D - A
print(L)

# 关联矩阵
B = nx.incidence_matrix(G)
print(B)

print(B.todense())
```
